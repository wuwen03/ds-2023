```python
import pickle
embeddings={}
embeddings=pickle.load(open("./embeddings.pkl",'rb'))
```

```python
import numpy as np
from sklearn.cluster import KMeans,BisectingKMeans,DBSCAN
keys=list(embeddings.keys())
values=np.array(list(embeddings.values()),dtype=float)

# n_clusters=15
model=DBSCAN(min_samples=5,eps=3)
model=model.fit(values)

# pred=model.predict(values)
labels=model.labels_
print(len(list(labels)))
n_clusters=len(set(labels))-(1 if -1 in labels else 0)
n_noise_=list(labels).count(-1)
print(n_clusters,n_noise_)
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
val=TSNE(n_components=2).fit_transform(values)
unique_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[model.core_sample_indices_] = True
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
 # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = val[class_member_mask & core_samples_mask]
    plt.scatter(
        xy[:, 0],
        xy[:, 1],
        marker="o",
        # markerfacecolor=tuple(col),
        # markeredgecolor="k",
        # markersize=6,
        alpha=0.7
    )

xy = val[class_member_mask & ~core_samples_mask]
plt.scatter(
    xy[:, 0],
    xy[:, 1],
    marker="o",
    # markerfacecolor=tuple(col),
    # markeredgecolor="k",
    # markersize=6,
    c='black',
    alpha=0.1
)

plt.title(f"Estimated number of clusters: {n_clusters}")
plt.show()
```
